```python
import pandas as pd
import matplotlib.pyplot as plt    
import seaborn as sns
y1=sns.load_dataset('penguins')
y1
```

```python
sns.barplot(x='species',y='bill_depth_mm',data=y1,hue="sex",ci=100  )     #sns.barplot(x='bill_depth_mm',y='species',data=y1)
plt.grid()
plt.show()   
                                                                           #order_1=(x='Chinstrap species','Gentoo','Adelie')
                                                                           #order=order_1,hue_order=['Female','Male']
```

```python
sns.barplot(x='body_mass_g',y='bill_depth_mm',data=y1,hue="sex",orient='v',palette='icefire',saturation=1000,errcolor='red',errwidth=3,capsize=0.5,dodge=False
            ,alpha=0.5 )    
 #orientation parameter is only used for two numerical data types
 #dodge=True/False is used to separate the value of hue
plt.grid()
plt.show()   
```

```python
import matplotlib.pyplot as plt    
import seaborn as sns
import pandas as pd
sns.displot(y1['bill_length_mm'],bins=[35,40,45,50,55,60],kde=True,rug= True,color='g')#log_scale=True
plt.show()
```

```python
import matplotlib.pyplot as plt    
import seaborn as sns
import pandas as pd
m={'Male':"*",'Female':'o'}
sns.scatterplot(x='body_mass_g',y='bill_depth_mm',data=y1,hue='sex',size='sex',sizes=(200,100),palette='icefire',markers=m)
plt.show()
```

```python
import matplotlib.pyplot as plt    
import seaborn as sns
import pandas as pd
import numpy as np
var=sns.load_dataset('tips')
var 

var=sns.violinplot(x='time',y='total_bill',data=var,hue='time',linewidth=2,palette='Dark2',order=['Dinner','Lunch'],split=True,scale='count',inner='stick')
plt.show()
```

```python
import matplotlib.pyplot as plt    
import seaborn as sns
import pandas as pd
import numpy as np
var=sns.load_dataset('tips')
var
sns.pairplot(var,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'],palette='BuGn',kind='kde',markers=['o','*'])
```
